Heatmat with Seaborn

1. Bibliotecas

```python
import numpy as np
import matplotlib as plt
import seaborn as sns
```

2.  Dados

```python
data = np.random.rand(10, 10)
data
```

3. Plot

```python
sns.heatmap(data=data,
            cmap='coolwarm',
            annot=True,
            fmt='.2f')
```
